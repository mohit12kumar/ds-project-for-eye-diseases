```python
# ============================================================
# Cell 7 : CLAHE Preprocessing
# RFMiD Multi-Class Retinal Disease Classification
# ============================================================

# ------------------------------------------------------------
# CLAHE Function
# ------------------------------------------------------------

def apply_clahe(image):
    """
    Apply CLAHE to a retinal RGB image.

    Parameters:
        image : PIL.Image

    Returns:
        Enhanced PIL.Image
    """

    # Convert PIL Image to NumPy Array
    image = np.array(image)

    # Convert RGB -> LAB Color Space
    lab = cv2.cvtColor(
        image,
        cv2.COLOR_RGB2LAB
    )

    # Split LAB Channels
    l, a, b = cv2.split(lab)

    # Create CLAHE Object
    clahe = cv2.createCLAHE(
        clipLimit=2.0,
        tileGridSize=(8, 8)
    )

    # Apply CLAHE on Lightness Channel
    l = clahe.apply(l)

    # Merge Channels
    enhanced_lab = cv2.merge((l, a, b))

    # Convert LAB -> RGB
    enhanced_image = cv2.cvtColor(
        enhanced_lab,
        cv2.COLOR_LAB2RGB
    )

    # Convert back to PIL Image
    enhanced_image = Image.fromarray(enhanced_image)

    return enhanced_image

print("=" * 60)
print("CLAHE Function Created Successfully")
print("=" * 60)
```

```python
# ============================================================
# Image Transformations
# RFMiD Multi-Class Retinal Disease Classification
# ============================================================

# ------------------------------------------------------------
# Training Transform
# ------------------------------------------------------------

train_transform = transforms.Compose([

    # Resize image for EfficientNet-B3
    transforms.Resize(
        (IMAGE_SIZE, IMAGE_SIZE)
    ),

    # Random horizontal flip
    transforms.RandomHorizontalFlip(
        p=0.5
    ),

    # Random vertical flip
    transforms.RandomVerticalFlip(
        p=0.3
    ),

    # Small rotation
    transforms.RandomRotation(
        degrees=15
    ),

    # Small affine transformation
    transforms.RandomAffine(
        degrees=0,
        translate=(0.05, 0.05),
        scale=(0.95, 1.05)
    ),

    # Brightness & Contrast
    transforms.ColorJitter(
        brightness=0.15,
        contrast=0.15,
        saturation=0.10,
        hue=0.02
    ),

    # Convert PIL Image → Tensor
    transforms.ToTensor(),

    # Normalize using ImageNet statistics
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    ),

    # Random Erasing
    transforms.RandomErasing(
        p=0.25,
        scale=(0.02, 0.10),
        ratio=(0.3, 3.3)
    )
])

# ------------------------------------------------------------
# Validation / Testing Transform
# ------------------------------------------------------------

val_transform = transforms.Compose([

    transforms.Resize(
        (IMAGE_SIZE, IMAGE_SIZE)
    ),

    transforms.ToTensor(),

    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

# ------------------------------------------------------------
# Test Time Augmentation (TTA)
# ------------------------------------------------------------

tta_transform = transforms.Compose([

    transforms.Resize(
        (IMAGE_SIZE, IMAGE_SIZE)
    ),

    transforms.RandomHorizontalFlip(
        p=1.0
    ),

    transforms.ToTensor(),

    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

print("=" * 60)
print("Image Transformations Created Successfully")
print("=" * 60)
```

```python
# ============================================================
# Cell 9 : Custom RFMiD Dataset
# Multi-Label Classification
# ============================================================

class RFMiDDataset(Dataset):

    def __init__(
        self,
        dataframe,
        image_dir,
        transform=None,
        apply_clahe_flag=True
    ):

        self.df = dataframe.reset_index(drop=True)

        self.image_dir = image_dir

        self.transform = transform

        self.apply_clahe = apply_clahe_flag

    def __len__(self):

        return len(self.df)

    def __getitem__(self, idx):

        # -----------------------------------------
        # Image ID
        # -----------------------------------------

        image_id = str(
            self.df.loc[idx, "ID"]
        )

        image_path = os.path.join(
            self.image_dir,
            image_id + ".png"
        )

        # -----------------------------------------
        # Image Exists?
        # -----------------------------------------

        if not os.path.exists(image_path):

            raise FileNotFoundError(
                f"Image Not Found : {image_path}"
            )

        # -----------------------------------------
        # Read Image
        # -----------------------------------------

        image = Image.open(
            image_path
        ).convert("RGB")

        # -----------------------------------------
        # CLAHE
        # -----------------------------------------

        if self.apply_clahe:

            image = apply_clahe(image)

        # -----------------------------------------
        # Transform
        # -----------------------------------------

        if self.transform:

            image = self.transform(image)

        # -----------------------------------------
        # Multi-Label Target
        # -----------------------------------------

        label = torch.tensor(

            self.df.loc[idx, CLASS_NAMES].values.astype(np.float32)

        )

        return image, label
```

```python
# ============================================================
# Cell 10 : Create Dataset & DataLoader
# RFMiD Multi-Label Classification
# ============================================================

# ------------------------------------------------------------
# Create Datasets
# ------------------------------------------------------------

train_dataset = RFMiDDataset(
    dataframe=train_df,
    image_dir=TRAIN_IMAGE_DIR,
    transform=train_transform,
    apply_clahe_flag=True
)

test_dataset = RFMiDDataset(
    dataframe=test_df,
    image_dir=TEST_IMAGE_DIR,
    transform=val_transform,
    apply_clahe_flag=True
)

eval_dataset = RFMiDDataset(
    dataframe=eval_df,
    image_dir=EVALUATION_IMAGE_DIR,
    transform=val_transform,
    apply_clahe_flag=True
)

# ------------------------------------------------------------
# Create DataLoaders
# ------------------------------------------------------------

train_loader = DataLoader(
    train_dataset,
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=0,          # Best for Google Colab
    pin_memory=torch.cuda.is_available()
)

test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=0,
    pin_memory=torch.cuda.is_available()
)

eval_loader = DataLoader(
    eval_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=0,
    pin_memory=torch.cuda.is_available()
)

# ------------------------------------------------------------
# Dataset Summary
# ------------------------------------------------------------

print("=" * 60)
print("Dataset Summary")
print("=" * 60)

print(f"Training Images    : {len(train_dataset)}")
print(f"Testing Images     : {len(test_dataset)}")
print(f"Evaluation Images  : {len(eval_dataset)}")

print()

print(f"Training Batches   : {len(train_loader)}")
print(f"Testing Batches    : {len(test_loader)}")
print(f"Evaluation Batches : {len(eval_loader)}")

print("=" * 60)
```

```python
# ============================================================
# Cell 11 : EfficientNet-B3 Model
# RFMiD Multi-Label Classification
# ============================================================

class EfficientNetB3RFMiD(nn.Module):

    def __init__(self, num_classes):

        super().__init__()

        # ----------------------------------------------------
        # Load Pretrained EfficientNet-B3
        # ----------------------------------------------------

        self.backbone = timm.create_model(
            "efficientnet_b3",
            pretrained=True,
            num_classes=0          # Remove original classifier
        )

        # ----------------------------------------------------
        # Feature Dimension
        # ----------------------------------------------------

        in_features = self.backbone.num_features

        # ----------------------------------------------------
        # Classification Head
        # ----------------------------------------------------

        self.classifier = nn.Sequential(

            nn.Dropout(0.30),

            nn.Linear(
                in_features,
                1024
            ),

            nn.ReLU(inplace=True),

            nn.BatchNorm1d(1024),

            nn.Dropout(0.30),

            nn.Linear(
                1024,
                512
            ),

            nn.ReLU(inplace=True),

            nn.BatchNorm1d(512),

            nn.Dropout(0.20),

            nn.Linear(
                512,
                num_classes
            )

        )

    def forward(self, x):

        x = self.backbone(x)

        x = self.classifier(x)

        return x


# ============================================================
# Create Model
# ============================================================

model = EfficientNetB3RFMiD(
    num_classes=NUM_CLASSES
)

model = model.to(DEVICE)

print("=" * 60)
print("EfficientNet-B3 Model Created Successfully")
print("=" * 60)
```

```python
# ============================================================
# Cell 12 : Loss Function, Optimizer & Scheduler
# RFMiD Multi-Label Classification
# ============================================================

# ------------------------------------------------------------
# Calculate Positive Class Weights
# ------------------------------------------------------------

labels = train_df[CLASS_NAMES].values.astype(np.float32)

positive = labels.sum(axis=0)

negative = len(train_df) - positive

# Prevent division by zero
positive = np.clip(positive, 1, None)

# pos_weight = Negative / Positive
pos_weight = negative / positive

pos_weight = torch.tensor(
    pos_weight,
    dtype=torch.float32
).to(DEVICE)

print("=" * 60)
print("Positive Class Weights")
print("=" * 60)

print(pos_weight)

# ------------------------------------------------------------
# Loss Function
# ------------------------------------------------------------

criterion = nn.BCEWithLogitsLoss(
    pos_weight=pos_weight
)

# ------------------------------------------------------------
# Optimizer
# ------------------------------------------------------------

optimizer = optim.AdamW(

    model.parameters(),

    lr=LEARNING_RATE,

    weight_decay=WEIGHT_DECAY

)

# ------------------------------------------------------------
# Learning Rate Scheduler
# ------------------------------------------------------------

scheduler = ReduceLROnPlateau(

    optimizer,

    mode="min",

    factor=LR_FACTOR,

    patience=LR_PATIENCE,

    min_lr=MIN_LR

)

# ------------------------------------------------------------
# Mixed Precision
# ------------------------------------------------------------

scaler = GradScaler(
    enabled=USE_AMP
)

print("\n" + "=" * 60)
print("Training Components Initialized Successfully")
print("=" * 60)

print(f"Loss Function      : BCEWithLogitsLoss")
print(f"Optimizer          : AdamW")
print(f"Scheduler          : ReduceLROnPlateau")
print(f"Mixed Precision    : {USE_AMP}")
print("=" * 60)
```

```python
# ============================================================
# Cell 13 : Training Function
# RFMiD Multi-Label Classification
# ============================================================

def train_one_epoch():

    # --------------------------------------------------------
    # Training Mode
    # --------------------------------------------------------

    model.train()

    running_loss = 0.0

    total_correct = 0

    total_predictions = 0

    progress_bar = tqdm(

        train_loader,

        desc="Training",

        leave=False

    )

    # --------------------------------------------------------
    # Training Loop
    # --------------------------------------------------------

    for images, labels in progress_bar:

        images = images.to(
            DEVICE,
            non_blocking=True
        )

        labels = labels.to(
            DEVICE,
            non_blocking=True
        )

        optimizer.zero_grad()

        # ----------------------------------------------------
        # Mixed Precision
        # ----------------------------------------------------

        with autocast(enabled=USE_AMP):

            outputs = model(images)

            loss = criterion(
                outputs,
                labels
            )

        # ----------------------------------------------------
        # Backpropagation
        # ----------------------------------------------------

        scaler.scale(loss).backward()

        scaler.unscale_(optimizer)

        torch.nn.utils.clip_grad_norm_(

            model.parameters(),

            MAX_GRAD_NORM

        )

        scaler.step(optimizer)

        scaler.update()

        # ----------------------------------------------------
        # Statistics
        # ----------------------------------------------------

        running_loss += loss.item()

        probabilities = torch.sigmoid(outputs)

        predictions = (

            probabilities >= 0.5

        ).float()

        total_correct += (

            predictions == labels

        ).sum().item()

        total_predictions += labels.numel()

        progress_bar.set_postfix(

            Loss=f"{loss.item():.4f}"

        )

    # --------------------------------------------------------
    # Epoch Results
    # --------------------------------------------------------

    epoch_loss = running_loss / len(train_loader)

    epoch_accuracy = (

        total_correct /

        total_predictions

    ) * 100

    return epoch_loss, epoch_accuracy
```

```python
# ============================================================
# Cell 14 : Validation Function
# RFMiD Multi-Label Classification
# ============================================================

@torch.no_grad()

def validate():

    # --------------------------------------------------------
    # Evaluation Mode
    # --------------------------------------------------------

    model.eval()

    running_loss = 0.0

    total_correct = 0

    total_predictions = 0

    all_labels = []

    all_probs = []

    all_preds = []

    progress_bar = tqdm(

        test_loader,

        desc="Validation",

        leave=False

    )

    # --------------------------------------------------------
    # Validation Loop
    # --------------------------------------------------------

    for images, labels in progress_bar:

        images = images.to(
            DEVICE,
            non_blocking=True
        )

        labels = labels.to(
            DEVICE,
            non_blocking=True
        )

        # ----------------------------------------------------
        # Forward Pass
        # ----------------------------------------------------

        with autocast(enabled=USE_AMP):

            outputs = model(images)

            loss = criterion(
                outputs,
                labels
            )

        running_loss += loss.item()

        # ----------------------------------------------------
        # Convert Logits -> Probabilities
        # ----------------------------------------------------

        probabilities = torch.sigmoid(outputs)

        predictions = (

            probabilities >= 0.5

        ).float()

        # ----------------------------------------------------
        # Accuracy
        # ----------------------------------------------------

        total_correct += (

            predictions == labels

        ).sum().item()

        total_predictions += labels.numel()

        # ----------------------------------------------------
        # Save Results
        # ----------------------------------------------------

        all_labels.append(

            labels.cpu()

        )

        all_probs.append(

            probabilities.cpu()

        )

        all_preds.append(

            predictions.cpu()

        )

        progress_bar.set_postfix(

            Loss=f"{loss.item():.4f}"

        )

    # --------------------------------------------------------
    # Epoch Statistics
    # --------------------------------------------------------

    epoch_loss = running_loss / len(test_loader)

    epoch_accuracy = (

        total_correct /

        total_predictions

    ) * 100

    # --------------------------------------------------------
    # Convert Lists to Arrays
    # --------------------------------------------------------

    all_labels = torch.cat(

        all_labels

    ).numpy()

    all_probs = torch.cat(

        all_probs

    ).numpy()

    all_preds = torch.cat(

        all_preds

    ).numpy()

    return (

        epoch_loss,

        epoch_accuracy,

        all_labels,

        all_probs,

        all_preds

    )
```

```python
# ------------------------------------------------------------
# Training History
# ------------------------------------------------------------

train_losses = []
val_losses = []

train_accs = []
val_accs = []

precision_scores = []
recall_scores = []
f1_scores = []

# ------------------------------------------------------------
# Early Stopping
# ------------------------------------------------------------

best_loss = float("inf")

best_epoch = 0

early_stop_counter = 0

# ------------------------------------------------------------
# Start Training
# ------------------------------------------------------------

print("="*70)
print("Training Started")
print("="*70)

training_start = time.time()

for epoch in range(EPOCHS):

    print(f"\nEpoch [{epoch+1}/{EPOCHS}]")
    print("-"*70)

    # --------------------------------------------------------
    # Train
    # --------------------------------------------------------

    train_loss, train_acc = train_one_epoch()

    # --------------------------------------------------------
    # Validation
    # --------------------------------------------------------

    val_loss, val_acc, labels, probs, preds = validate()

    # --------------------------------------------------------
    # Scheduler
    # --------------------------------------------------------

    scheduler.step(val_loss)

    # --------------------------------------------------------
    # Precision Recall F1
    # --------------------------------------------------------

    precision = precision_score(
        labels,
        preds,
        average="samples",
        zero_division=0
    )

    recall = recall_score(
        labels,
        preds,
        average="samples",
        zero_division=0
    )

    f1 = f1_score(
        labels,
        preds,
        average="samples",
        zero_division=0
    )

    # --------------------------------------------------------
    # Save History
    # --------------------------------------------------------

    train_losses.append(train_loss)
    val_losses.append(val_loss)

    train_accs.append(train_acc)
    val_accs.append(val_acc)

    precision_scores.append(precision)
    recall_scores.append(recall)
    f1_scores.append(f1)

    # --------------------------------------------------------
    # Print Results
    # --------------------------------------------------------

    print(f"Train Loss      : {train_loss:.4f}")
    print(f"Validation Loss : {val_loss:.4f}")

    print(f"Train Accuracy  : {train_acc:.2f}%")
    print(f"Validation Acc. : {val_acc:.2f}%")

    print(f"Precision       : {precision:.4f}")
    print(f"Recall          : {recall:.4f}")
    print(f"F1 Score        : {f1:.4f}")

    current_lr = optimizer.param_groups[0]["lr"]

    print(f"Learning Rate   : {current_lr:.7f}")

    # --------------------------------------------------------
    # Save Best Model
    # --------------------------------------------------------

    if val_loss < best_loss:

        best_loss = val_loss

        best_epoch = epoch + 1

        early_stop_counter = 0

        torch.save(
            model.state_dict(),
            BEST_MODEL_PATH
        )

        print("Best Model Saved")

    else:

        early_stop_counter += 1

        print(
            f"No Improvement ({early_stop_counter}/{EARLY_STOPPING_PATIENCE})"
        )

    # --------------------------------------------------------
    # Early Stopping
    # --------------------------------------------------------

    if early_stop_counter >= EARLY_STOPPING_PATIENCE:

        print("\nEarly Stopping Triggered")

        break

# ------------------------------------------------------------
# Training Finished
# ------------------------------------------------------------

training_end = time.time()

training_time = training_end - training_start

print("\n" + "="*70)

print("Training Completed")

print("="*70)

print(f"Best Epoch       : {best_epoch}")

print(f"Best Val Loss    : {best_loss:.4f}")

print(f"Training Time    : {training_time/60:.2f} Minutes")

print(f"Model Saved At   : {BEST_MODEL_PATH}")

print("="*70)
```

```python
# ============================================================
# Cell 22 : Test-Time Augmentation (TTA)
# RFMiD Multi-Label Classification
# ============================================================

# ------------------------------------------------------------
# TTA Transforms
# ------------------------------------------------------------

tta_transforms = [

    transforms.Compose([
        transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.485,0.456,0.406],
            std=[0.229,0.224,0.225]
        )
    ]),

    transforms.Compose([
        transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
        transforms.RandomHorizontalFlip(p=1.0),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.485,0.456,0.406],
            std=[0.229,0.224,0.225]
        )
    ]),

    transforms.Compose([
        transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
        transforms.RandomVerticalFlip(p=1.0),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.485,0.456,0.406],
            std=[0.229,0.224,0.225]
        )
    ]),

    transforms.Compose([
        transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
        transforms.RandomRotation(10),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.485,0.456,0.406],
            std=[0.229,0.224,0.225]
        )
    ])

]

# ============================================================
# TTA Prediction Function
# ============================================================

def tta_predict(image):

    model.eval()

    predictions = []

    with torch.no_grad():

        for transform in tta_transforms:

            img = transform(image)

            img = img.unsqueeze(0).to(DEVICE)

            output = model(img)

            prob = torch.sigmoid(output)

            predictions.append(prob.cpu().numpy())

    predictions = np.array(predictions)

    mean_prediction = predictions.mean(axis=0)[0]

    return mean_prediction
```
