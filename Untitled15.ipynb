```python
# ============================================================
# Cell 7 : CLAHE Preprocessing
# ============================================================

# -----------------------------
# CLAHE Function
# -----------------------------
def apply_clahe(image):

    # Convert PIL Image to NumPy array
    image = np.array(image)

    # Convert RGB to LAB color space
    lab = cv2.cvtColor(image, cv2.COLOR_RGB2LAB)

    # Split LAB channels
    l, a, b = cv2.split(lab)

    # Apply CLAHE on L channel
    clahe = cv2.createCLAHE(
        clipLimit=2.0,
        tileGridSize=(8, 8)
    )

    l = clahe.apply(l)

    # Merge channels
    lab = cv2.merge((l, a, b))

    # Convert LAB back to RGB
    image = cv2.cvtColor(lab, cv2.COLOR_LAB2RGB)

    # Convert NumPy array back to PIL Image
    image = Image.fromarray(image)

    return image

# ============================================================
# Display Before & After CLAHE
# ============================================================

# Select one sample image
image_id = train_df.iloc[0]["ID"]

image_path = os.path.join(
    TRAIN_IMAGE_DIR,
    f"{image_id}.png"
)

# Load image
original_image = Image.open(image_path).convert("RGB")

# Apply CLAHE
clahe_image = apply_clahe(original_image)

# Display results
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(original_image)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(clahe_image)
plt.title("CLAHE Image")
plt.axis("off")

plt.tight_layout()
plt.show()

print("=" * 60)
print("CLAHE Preprocessing Completed Successfully")
print("=" * 60)
```

```python
# ============================================================
# Cell 3 : Project Configuration
# ============================================================

# -----------------------------
# Random Seed
# -----------------------------
SEED = 42

# -----------------------------
# Image Size
# -----------------------------
IMAGE_SIZE = 384

# -----------------------------
# Training Parameters
# -----------------------------
BATCH_SIZE = 16
EPOCHS = 25
LEARNING_RATE = 0.0001
WEIGHT_DECAY = 1e-4

# -----------------------------
# Binary Classification
# -----------------------------
NUM_CLASSES = 2

# -----------------------------
# Device
# -----------------------------
DEVICE = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

# -----------------------------
# Set Random Seed
# -----------------------------
random.seed(SEED)
np.random.seed(SEED)

torch.manual_seed(SEED)

if torch.cuda.is_available():

    torch.cuda.manual_seed(SEED)
    torch.cuda.manual_seed_all(SEED)

# -----------------------------
# Display Configuration
# -----------------------------
print("=" * 60)

print("Project Configuration")

print("=" * 60)

print(f"Image Size      : {IMAGE_SIZE}")
print(f"Batch Size      : {BATCH_SIZE}")
print(f"Epochs          : {EPOCHS}")
print(f"Learning Rate   : {LEARNING_RATE}")
print(f"Weight Decay    : {WEIGHT_DECAY}")
print(f"Number of Class : {NUM_CLASSES}")
print(f"Device          : {DEVICE}")

print("=" * 60)
```

```python
# ============================================================
# Cell 8 : Image Transformations
# ============================================================

# -----------------------------
# Training Transform
# -----------------------------
train_transform = transforms.Compose([

    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),

    transforms.RandomHorizontalFlip(p=0.5),

    transforms.RandomRotation(10),

    transforms.ToTensor(),

    transforms.Normalize(

        mean=[0.485, 0.456, 0.406],

        std=[0.229, 0.224, 0.225]

    )

])

# -----------------------------
# Validation / Testing Transform
# -----------------------------
val_transform = transforms.Compose([

    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),

    transforms.ToTensor(),

    transforms.Normalize(

        mean=[0.485, 0.456, 0.406],

        std=[0.229, 0.224, 0.225]

    )

])

print("=" * 60)

print("Image Transformations Created Successfully")

print("=" * 60)
```

```python
# ============================================================
# Cell 9 : Custom Dataset
# ============================================================

class RFMiDDataset(Dataset):

    def __init__(self, dataframe, image_dir, transform=None):

        self.dataframe = dataframe

        self.image_dir = image_dir

        self.transform = transform

    # --------------------------------------------------------
    # Total number of images
    # --------------------------------------------------------

    def __len__(self):

        return len(self.dataframe)

    # --------------------------------------------------------
    # Load one image and its label
    # --------------------------------------------------------

    def __getitem__(self, index):

        # Get Image ID
        image_id = self.dataframe.iloc[index]["ID"]

        # Create image path
        image_path = os.path.join(

            self.image_dir,

            f"{image_id}.png"

        )

        # Read image
        image = Image.open(image_path).convert("RGB")

        # Apply CLAHE
        image = apply_clahe(image)

        # Apply Transform
        if self.transform:

            image = self.transform(image)

        # Binary Label
        label = int(

            self.dataframe.iloc[index]["Disease_Risk"]

        )

        label = torch.tensor(

            label,

            dtype=torch.long

        )

        return image, label
```

```python
# ============================================================
# Create Dataset
# ============================================================

train_dataset = RFMiDDataset(

    dataframe=train_df,

    image_dir=TRAIN_IMAGE_DIR,

    transform=train_transform

)

test_dataset = RFMiDDataset(

    dataframe=test_df,

    image_dir=TEST_IMAGE_DIR,

    transform=val_transform

)

print("=" * 60)

print("Dataset Created Successfully")

print("=" * 60)

print("Training Images :", len(train_dataset))

print("Testing Images  :", len(test_dataset))
```

```python
# ============================================================
# Cell 10 : DataLoader
# ============================================================

# -----------------------------
# Training DataLoader
# -----------------------------
train_loader = DataLoader(

    train_dataset,

    batch_size=BATCH_SIZE,

    shuffle=True,

    num_workers=2,

    pin_memory=True

)

# -----------------------------
# Testing DataLoader
# -----------------------------
test_loader = DataLoader(

    test_dataset,

    batch_size=BATCH_SIZE,

    shuffle=False,

    num_workers=2,

    pin_memory=True

)

print("=" * 60)

print("DataLoaders Created Successfully")

print("=" * 60)

print("Training Batches :", len(train_loader))

print("Testing Batches  :", len(test_loader))
```

```python
# ============================================================
# Cell 11 : EfficientNet-B3 Model
# ============================================================

class EyeDiseaseModel(nn.Module):

    def __init__(self):

        super(EyeDiseaseModel, self).__init__()

        # ------------------------------------------
        # Load Pretrained EfficientNet-B3
        # ------------------------------------------

        self.backbone = timm.create_model(

            "efficientnet_b3",

            pretrained=True,

            num_classes=0

        )

        # ------------------------------------------
        # Number of Features
        # ------------------------------------------

        in_features = self.backbone.num_features

        # ------------------------------------------
        # Classification Layer
        # ------------------------------------------

        self.classifier = nn.Sequential(

            nn.Dropout(0.3),

            nn.Linear(

                in_features,

                NUM_CLASSES

            )

        )

    # ------------------------------------------
    # Forward Pass
    # ------------------------------------------

    def forward(self, x):

        x = self.backbone(x)

        x = self.classifier(x)

        return x


# ============================================================
# Create Model
# ============================================================

model = EyeDiseaseModel().to(DEVICE)

print("=" * 60)

print("EfficientNet-B3 Model Created Successfully")

print("=" * 60)
```

```python
# ============================================================
# Cell 12 : Loss Function, Optimizer & Scheduler
# ============================================================

# -----------------------------
# Calculate Class Weights
# -----------------------------
class_weights = compute_class_weight(

    class_weight="balanced",

    classes=np.unique(train_df["Disease_Risk"]),

    y=train_df["Disease_Risk"]

)

class_weights = torch.tensor(

    class_weights,

    dtype=torch.float

).to(DEVICE)

print("Class Weights :", class_weights)

# -----------------------------
# Loss Function
# -----------------------------
criterion = nn.CrossEntropyLoss(

    weight=class_weights

)

# -----------------------------
# Optimizer
# -----------------------------
optimizer = optim.AdamW(

    model.parameters(),

    lr=LEARNING_RATE,

    weight_decay=WEIGHT_DECAY

)

# -----------------------------
# Learning Rate Scheduler
# -----------------------------
scheduler = optim.lr_scheduler.ReduceLROnPlateau(

    optimizer,

    mode="min",

    factor=0.1,

    patience=3

)

# -----------------------------
# Display Configuration
# -----------------------------
print("=" * 60)

print("Training Configuration")

print("=" * 60)

print("Loss Function : CrossEntropyLoss")

print("Optimizer     : AdamW")

print("Scheduler     : ReduceLROnPlateau")

print("=" * 60)
```

```python
# ============================================================
# Cell 13 : Training Function
# ============================================================

def train_one_epoch(model, dataloader, criterion, optimizer):

    # Set model to training mode
    model.train()

    running_loss = 0.0
    correct = 0
    total = 0

    # Loop through all batches
    for images, labels in tqdm(dataloader, desc="Training"):

        # Move data to GPU/CPU
        images = images.to(DEVICE)
        labels = labels.to(DEVICE)

        # Clear previous gradients
        optimizer.zero_grad()

        # Forward pass
        outputs = model(images)

        # Calculate loss
        loss = criterion(outputs, labels)

        # Backpropagation
        loss.backward()

        # Update weights
        optimizer.step()

        # Add batch loss
        running_loss += loss.item()

        # Predicted class
        _, predicted = torch.max(outputs, 1)

        # Count correct predictions
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    # Average loss
    epoch_loss = running_loss / len(dataloader)

    # Accuracy
    epoch_accuracy = 100 * correct / total

    return epoch_loss, epoch_accuracy
```

```python
# ============================================================
# Cell 14 : Validation Function
# ============================================================

def validate(model, dataloader, criterion):

    # Set model to evaluation mode
    model.eval()

    running_loss = 0.0

    all_labels = []
    all_predictions = []
    all_probabilities = []

    # Disable gradient calculation
    with torch.no_grad():

        for images, labels in tqdm(dataloader, desc="Validation"):

            # Move data to device
            images = images.to(DEVICE)
            labels = labels.to(DEVICE)

            # Forward pass
            outputs = model(images)

            # Calculate loss
            loss = criterion(outputs, labels)

            running_loss += loss.item()

            # Convert scores to probabilities
            probabilities = torch.softmax(outputs, dim=1)

            # Predicted class
            _, predictions = torch.max(outputs, 1)

            # Store results
            all_labels.extend(labels.cpu().numpy())

            all_predictions.extend(predictions.cpu().numpy())

            all_probabilities.extend(
                probabilities[:,1].cpu().numpy()
            )

    # Average Validation Loss
    val_loss = running_loss / len(dataloader)

    # Evaluation Metrics
    accuracy = accuracy_score(
        all_labels,
        all_predictions
    )

    precision = precision_score(
        all_labels,
        all_predictions,
        zero_division=0
    )

    recall = recall_score(
        all_labels,
        all_predictions,
        zero_division=0
    )

    f1 = f1_score(
        all_labels,
        all_predictions,
        zero_division=0
    )

    roc_auc = roc_auc_score(
        all_labels,
        all_probabilities
    )

    return (

        val_loss,

        accuracy,

        precision,

        recall,

        f1,

        roc_auc

    )
```

```python
# ============================================================
# Cell 15 : Model Training with Early Stopping
# ============================================================

import torch

# -------------------------------------------------
# Variables
# -------------------------------------------------
best_accuracy = 0.0
best_val_loss = float("inf")
best_epoch = 0

patience = 5          # Stop after 5 epochs without improvement
counter = 0

train_losses = []
train_accuracies = []

val_losses = []
val_accuracies = []

print("=" * 70)
print("Training Started")
print("=" * 70)

for epoch in range(EPOCHS):

    print(f"\nEpoch [{epoch + 1}/{EPOCHS}]")

    # -------------------------------------------------
    # Training
    # -------------------------------------------------
    train_loss, train_acc = train_one_epoch(
        model,
        train_loader,
        criterion,
        optimizer
    )

    # -------------------------------------------------
    # Validation
    # -------------------------------------------------
    val_loss, val_acc, precision, recall, f1, roc_auc = validate(
        model,
        test_loader,
        criterion
    )

    # -------------------------------------------------
    # Learning Rate Scheduler
    # -------------------------------------------------
    scheduler.step(val_loss)

    # -------------------------------------------------
    # Save History
    # -------------------------------------------------
    train_losses.append(train_loss)
    train_accuracies.append(train_acc)

    val_losses.append(val_loss)
    val_accuracies.append(val_acc * 100)

    # -------------------------------------------------
    # Early Stopping + Save Best Model
    # -------------------------------------------------
    if val_loss < best_val_loss:

        best_val_loss = val_loss
        best_accuracy = val_acc
        best_epoch = epoch + 1
        counter = 0

        torch.save(
            model.state_dict(),
            "best_model.pth"
        )

        print(f"✅ New Best Model Saved (Epoch {best_epoch})")

    else:

        counter += 1

        print(
            f"⚠️ Validation loss did not improve "
            f"({counter}/{patience})"
        )

    # -------------------------------------------------
    # Print Metrics
    # -------------------------------------------------
    print(f"Train Loss      : {train_loss:.4f}")
    print(f"Validation Loss : {val_loss:.4f}")

    print(f"Train Accuracy  : {train_acc:.2f}%")
    print(f"Validation Acc. : {val_acc * 100:.2f}%")

    # print(f"Precision       : {precision:.4f}")
    # print(f"Recall          : {recall:.4f}")
    # print(f"F1 Score        : {f1:.4f}")
    # print(f"ROC AUC         : {roc_auc:.4f}")

    print(f"Best Epoch So Far : {best_epoch}")
    print(f"Best Val Loss     : {best_val_loss:.4f}")
    print(f"Best Val Accuracy : {best_accuracy * 100:.2f}%")

    # -------------------------------------------------
    # Early Stopping Check
    # -------------------------------------------------
    if counter >= patience:

        print("\n" + "=" * 70)
        print("🛑 EARLY STOPPING TRIGGERED")
        print("=" * 70)
        print(f"No improvement for {patience} consecutive epochs.")
        print(f"Best Model Saved at Epoch : {best_epoch}")
        print(f"Best Validation Loss      : {best_val_loss:.4f}")
        print(f"Best Validation Accuracy  : {best_accuracy * 100:.2f}%")
        print("=" * 70)

        break

# -------------------------------------------------
# Training Completed
# -------------------------------------------------
print("\n" + "=" * 70)
print("Training Completed Successfully")
print("=" * 70)

print(f"🏆 Best Epoch               : {best_epoch}")
print(f"🏆 Best Validation Accuracy : {best_accuracy * 100:.2f}%")
print(f"🏆 Best Validation Loss     : {best_val_loss:.4f}")

print("=" * 70)

# -------------------------------------------------
# Load Best Model
# -------------------------------------------------
model.load_state_dict(torch.load("best_model.pth"))

print(f"✅ Best model from Epoch {best_epoch} loaded successfully.")
```

```python
# ============================================================
# Cell 16 : Load Best Model & Final Evaluation
# ============================================================

# -----------------------------
# Load Best Model
# -----------------------------
model.load_state_dict(torch.load("best_model.pth"))

model.to(DEVICE)

model.eval()

print("=" * 60)
print("Best Model Loaded Successfully")
print("=" * 60)

# -----------------------------
# Lists to Store Results
# -----------------------------
all_labels = []
all_predictions = []
all_probabilities = []

# -----------------------------
# Model Evaluation
# -----------------------------
with torch.no_grad():

    for images, labels in tqdm(test_loader, desc="Testing"):

        images = images.to(DEVICE)
        labels = labels.to(DEVICE)

        outputs = model(images)

        probabilities = torch.softmax(outputs, dim=1)

        _, predictions = torch.max(outputs, 1)

        all_labels.extend(labels.cpu().numpy())

        all_predictions.extend(predictions.cpu().numpy())

        all_probabilities.extend(
            probabilities[:, 1].cpu().numpy()
        )

# -----------------------------
# Calculate Metrics
# -----------------------------
accuracy = accuracy_score(
    all_labels,
    all_predictions
)

precision = precision_score(
    all_labels,
    all_predictions,
    zero_division=0
)

recall = recall_score(
    all_labels,
    all_predictions,
    zero_division=0
)

f1 = f1_score(
    all_labels,
    all_predictions,
    zero_division=0
)

roc_auc = roc_auc_score(
    all_labels,
    all_probabilities
)

# -----------------------------
# Print Results
# -----------------------------
print("\n" + "=" * 60)
print("Final Model Performance")
print("=" * 60)

print(f"Accuracy  : {accuracy*100:.2f}%")
print(f"Precision : {precision:.4f}")
print(f"Recall    : {recall:.4f}")
print(f"F1 Score  : {f1:.4f}")
print(f"ROC AUC   : {roc_auc:.4f}")

print("=" * 60)
```

```python
# ============================================================
# Cell 21 : Single Image Prediction
# ============================================================

import time

# ------------------------------------------------------------
# Predict Function
# ------------------------------------------------------------

def predict_image(image_path):

    # Load image
    image = Image.open(image_path).convert("RGB")

    # Apply CLAHE
    image = apply_clahe(image)

    # Apply validation transform
    input_image = val_transform(image)

    # Add batch dimension
    input_image = input_image.unsqueeze(0).to(DEVICE)

    # Prediction
    model.eval()

    start_time = time.time()

    with torch.no_grad():

        outputs = model(input_image)

        probabilities = torch.softmax(outputs, dim=1)

        confidence, prediction = torch.max(probabilities, 1)

    end_time = time.time()

    # Class Names
    class_names = ["Healthy", "Disease"]

    predicted_class = class_names[prediction.item()]

    confidence = confidence.item() * 100

    inference_time = (end_time - start_time) * 1000

    # Display Image
    plt.figure(figsize=(6,6))

    plt.imshow(image)

    plt.title(
        f"Prediction : {predicted_class}\n"
        f"Confidence : {confidence:.2f}%"
    )

    plt.axis("off")

    plt.show()

    # Print Result
    print("=" * 60)

    print(f"Prediction      : {predicted_class}")

    print(f"Confidence      : {confidence:.2f}%")

    print(f"Inference Time  : {inference_time:.2f} ms")

    print("=" * 60)

    return predicted_class, confidence
```

```python
# ============================================================
# Cell 22 : Multiple Image Prediction
# ============================================================

results = []

print("=" * 60)
print("Predicting Test Images...")
print("=" * 60)

model.eval()

with torch.no_grad():

    for _, row in tqdm(test_df.iterrows(),
                       total=len(test_df),
                       desc="Testing"):

        # Image ID
        image_id = row["ID"]

        # Actual Label
        actual_label = row["Disease_Risk"]

        # Image Path
        image_path = os.path.join(
            TEST_IMAGE_DIR,
            f"{image_id}.png"
        )

        # Load Image
        image = Image.open(image_path).convert("RGB")

        # Apply CLAHE
        image = apply_clahe(image)

        # Transform
        image_tensor = val_transform(image)

        image_tensor = image_tensor.unsqueeze(0).to(DEVICE)

        # Prediction
        outputs = model(image_tensor)

        probabilities = torch.softmax(outputs, dim=1)

        confidence, prediction = torch.max(probabilities, 1)

        predicted_label = prediction.item()

        confidence = confidence.item() * 100

        # Save Result
        results.append({

            "Image_ID": image_id,

            "Actual_Label": actual_label,

            "Predicted_Label": predicted_label,

            "Confidence (%)": round(confidence,2)

        })

# ------------------------------------------------------------
# Convert to DataFrame
# ------------------------------------------------------------

results_df = pd.DataFrame(results)

# Convert labels to names
label_names = {
    0: "Healthy",
    1: "Disease"
}

results_df["Actual_Label"] = results_df["Actual_Label"].map(label_names)

results_df["Predicted_Label"] = results_df["Predicted_Label"].map(label_names)

# Correct Prediction
results_df["Correct"] = (

    results_df["Actual_Label"]

    ==

    results_df["Predicted_Label"]

)

# ------------------------------------------------------------
# Display Results
# ------------------------------------------------------------

display(results_df.head())

# ------------------------------------------------------------
# Save Results
# ------------------------------------------------------------

results_df.to_csv(

    "prediction_results.csv",

    index=False

)

print("\nPrediction results saved as 'prediction_results.csv'")

print("=" * 60)
```
